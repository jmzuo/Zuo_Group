```python
cr.list_all_builtin_crystals()
```

```python
bn = cr.from_builtin('Boron_Tetra')
```

```python
print(bn)
```

```python
_, bndp = bn.generateDP()
```

```python
bndp.plot()
```

```python
from pyemaps import EMC as ec
```

```python
emc = ec(dict(zone=(1,1,0), defl=(0.0,0.0), cl = 2000, vt =300, tilt=(1.0,0.0)))
```

```python
_, bndp1 = bn.generateDP(mode=1, dsize=0, em_controls = emc)
```

```python
bndp1.plot()
```

```python
bnpd = bn.generatePowder()
```

```python
bn.plotPowder(bnpd)
```

```python
bnpd = bn.generatePowder(kv=300, t2max = 0.05, 
            smax = 1.5, 
            eta = 1.0, 
            gamma = 0.001, 
            absp = 0, 
            bg = False,
            bamp = 0.35, 
            bgamma = 0.001, 
            bmfact = 0.02)
```

```python
bn.plotPowder(bnpd)
```
